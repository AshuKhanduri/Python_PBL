```python
import numpy as np
import pandas as pd

np.random.seed(42)
n = 10000
current_year = 2025

# -------------------------
# CORE STRUCTURAL FEATURES
# -------------------------

area = np.random.gamma(2.5, 900, n)
area = np.clip(area, 500, 8000)

floors = np.where(area > 3000,
                  np.random.randint(2, 5, n),
                  np.random.randint(1, 3, n))

bedrooms = np.where(area > 2500,
                    np.random.randint(3, 7, n),
                    np.random.randint(1, 4, n))

bathrooms = np.clip(bedrooms - np.random.randint(0, 2, n), 1, 5)

lot_size = area + np.random.uniform(200, 2000, n)

year_built = np.random.randint(1970, 2025, n)
property_age = current_year - year_built

# -------------------------
# LOCATION & ENVIRONMENT
# -------------------------

distance_city = np.random.uniform(0, 40, n)

neighborhood_quality = np.random.randint(1, 11, n)

crime_rate = np.clip(100 - neighborhood_quality * 8 + 
                     np.random.normal(0, 5, n), 0, 100)

school_rating = np.random.randint(1, 11, n)

hospital_proximity = np.random.uniform(0.5, 20, n)
shopping_proximity = np.random.uniform(0.5, 25, n)

public_transport = np.random.randint(1, 11, n)

green_space = np.clip(neighborhood_quality * 8 + 
                      np.random.normal(0, 10, n), 0, 100)

flood_risk = np.clip(100 - green_space + 
                     np.random.normal(0, 10, n), 0, 100)

noise_pollution = np.clip(70 - green_space * 0.3 + 
                          np.random.normal(0, 5, n), 20, 100)

# -------------------------
# INFRASTRUCTURE & UTILITIES
# -------------------------

construction_quality = np.random.randint(1, 11, n)

energy_efficiency = np.random.randint(1, 11, n)

water_supply = np.random.randint(1, 11, n)
electricity_supply = np.random.randint(1, 11, n)
internet_score = np.random.randint(1, 11, n)

parking = np.random.choice([0, 1], n, p=[0.3, 0.7])

renovation_status = np.random.choice(["Yes", "No"], n, p=[0.4, 0.6])

property_type = np.random.choice(
    ["Apartment", "Villa", "Independent House"],
    n, p=[0.5, 0.2, 0.3]
)

# -------------------------
# PRICE FORMULA
# -------------------------

base_price = (
    area * 4000
    + bedrooms * 70000
    + bathrooms * 50000
    + floors * 80000
    + neighborhood_quality * 50000
    + school_rating * 30000
    + construction_quality * 60000
    + energy_efficiency * 40000
    + public_transport * 30000
    + green_space * 5000
    - crime_rate * 15000
    - distance_city * 20000
    - property_age * 10000
    - flood_risk * 8000
    - noise_pollution * 3000
)

type_bonus = np.where(property_type == "Villa", 1500000,
              np.where(property_type == "Independent House", 900000, 500000))

renovation_bonus = np.where(renovation_status == "Yes", 400000, 0)

parking_bonus = parking * 300000

noise = np.random.normal(0, 800000, n)

price = base_price + type_bonus + renovation_bonus + parking_bonus + noise
price = np.clip(price, 500000, None)

# -------------------------
# CREATE DATAFRAME
# -------------------------

df = pd.DataFrame({
    "Property_Area": area,
    "Bedrooms": bedrooms,
    "Bathrooms": bathrooms,
    "Floors": floors,
    "Year_Built": year_built,
    "Property_Age": property_age,
    "Renovation_Status": renovation_status,
    "Lot_Size": lot_size,
    "Distance_to_City_Center": distance_city,
    "Neighborhood_Quality_Score": neighborhood_quality,
    "Crime_Rate_Index": crime_rate,
    "Nearby_School_Rating": school_rating,
    "Hospital_Proximity_km": hospital_proximity,
    "Shopping_Center_Proximity_km": shopping_proximity,
    "Public_Transport_Accessibility": public_transport,
    "Parking_Availability": parking,
    "Property_Type": property_type,
    "Construction_Quality_Rating": construction_quality,
    "Energy_Efficiency_Score": energy_efficiency,
    "Water_Supply_Reliability": water_supply,
    "Electricity_Supply_Reliability": electricity_supply,
    "Internet_Availability_Score": internet_score,
    "Green_Space_Availability_Index": green_space,
    "Flood_Risk_Index": flood_risk,
    "Noise_Pollution_Level": noise_pollution,
    "Price": price
})

# Add 5% missing values
for col in ["Lot_Size", "Energy_Efficiency_Score",
            "Green_Space_Availability_Index",
            "Nearby_School_Rating"]:
    df.loc[df.sample(frac=0.05).index, col] = np.nan

# Add some luxury outliers
outliers = np.random.choice(df.index, 15, replace=False)
df.loc[outliers, "Property_Area"] *= 2
df.loc[outliers, "Price"] *= 2

df.to_csv("house_data.csv", index=False)

print("Dataset generated successfully!")
print(df.head())
```
